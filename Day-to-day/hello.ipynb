```python
chunks = [] # List to keep filtered chunk dataframes.

for chunked_cars_df in pd.read_csv("cars.csv", 
                                               header=0, 
                                               sep=",", 
                                               chunksize=100):
    # Filter for 6 cylinder cars and append dataframe to the list
    chunks.append(chunked_cars_df.query("Cylinders==6").copy())

print(len(chunks))
```

```python
for chunk in chunks:
    print(chunk.shape)
```

```python
cars = pd.concat(chunks)
cars.shape
```
